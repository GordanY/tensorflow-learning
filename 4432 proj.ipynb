```
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt


from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import GradientBoostingRegressor
```

```
from google.colab import drive
drive.mount('/content/gdrive')

X_removed = pd.read_csv('/content/gdrive/My Drive/data_final.csv')
```

```
y_revenue_removed = np.log(X_removed['revenue'])

X_budget = X_removed['budget'].values.reshape(-1, 1)
X_budget = np.log(X_budget + 1.0)

X_removed['budget_zero'] = np.where(X_removed['budget'] == 0, 1, 0)
X_budget_zero = X_removed['budget_zero'].values.reshape(-1, 1)
```

```
# Processing Genres data
genres_prop = np.array([])
for i in range(len(X_removed)):
    temp = json.loads(X_removed['genres'][i])
    temp_2 = np.array([])
    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])
    genres_prop = np.append(genres_prop, temp_2)

# Get set of unique elements in movie's genres
genres_prop = list(set(genres_prop))

X_genres = np.array([])
for i in range(len(X_removed)):
    null_genres = [0]*len(genres_prop)
    temp = json.loads(X_removed['genres'][i])
    temp_2 = np.array([])

    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])

    for j in range(len(temp_2)):
        if temp_2[j] in genres_prop:
            null_genres[genres_prop.index(temp_2[j])] = 1
        else:
            null_genres[genres_prop.index(temp_2[j])] = 0
    X_genres = np.append(X_genres, null_genres)

# Genres data, 1 indicate the movie belongs to that genres' type
X_genres = X_genres.reshape(-1, len(genres_prop))
```

```
# Processing Keywords data
keywords_prop = np.array([])
for i in range(len(X_removed)):
    temp = json.loads(X_removed['keywords'][i])
    temp_2 = np.array([])
    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])
    keywords_prop = np.append(keywords_prop, temp_2)

# Get set of unique elements in movie's keywords
keywords_prop = list(set(keywords_prop))

X_keywords = np.array([])
for i in range(len(X_removed)):
    null_keywords = [0]*len(keywords_prop)
    temp = json.loads(X_removed['keywords'][i])
    temp_2 = np.array([])

    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])

    for j in range(len(temp_2)):
        if temp_2[j] in keywords_prop:
            null_keywords[keywords_prop.index(temp_2[j])] = 1
        else:
            null_keywords[keywords_prop.index(temp_2[j])] = 0
    X_keywords = np.append(X_keywords, null_keywords)

# Genres data, 1 indicate the movie belongs to that genres' type
X_keywords = X_keywords.reshape(-1, len(keywords_prop))
```

```
# Processing production_companies data
pro_company_prop = np.array([])
for i in range(len(X_removed)):
    temp = json.loads(X_removed['production_companies'][i])
    temp_2 = np.array([])
    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])
    pro_company_prop = np.append(pro_company_prop, temp_2)

# Get set of unique elements in movie's keywords
pro_company_prop = list(set(pro_company_prop))

X_prop_company = np.array([])
for i in range(len(X_removed)):
    null_keywords = [0]*len(pro_company_prop)
    temp = json.loads(X_removed['production_companies'][i])
    temp_2 = np.array([])

    for j in range(len(temp)):
        temp_2 = np.append(temp_2, temp[j]['name'])

    for j in range(len(temp_2)):
        if temp_2[j] in pro_company_prop:
            null_keywords[pro_company_prop.index(temp_2[j])] = 1
        else:
            null_keywords[pro_company_prop.index(temp_2[j])] = 0
    X_prop_company = np.append(X_prop_company, null_keywords)

# Genres data, 1 indicate the movie belongs to that genres' type
X_prop_company = X_prop_company.reshape(-1, len(pro_company_prop))
```

```
reg = LinearRegression()
```

```
X = np.concatenate((X_budget, X_budget_zero), axis = 1)
reg.fit(X, y_revenue_removed)
reg.score(X, y_revenue_removed)
```

```
X = np.concatenate((X_budget, X_budget_zero, X_genres), axis = 1)
reg.fit(X, y_revenue_removed)
reg.score(X, y_revenue_removed)
```

```
X = np.concatenate((X_budget, X_budget_zero, X_genres, X_keywords), axis = 1)
reg.fit(X, y_revenue_removed)
reg.score(X, y_revenue_removed)
```

```
X = np.concatenate((X_budget, X_budget_zero, X_genres, X_keywords, X_prop_company), axis = 1)
reg.fit(X, y_revenue_removed)
reg.score(X, y_revenue_removed)
```
